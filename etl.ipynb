## Tabla de valor bruto de producción en miles de dólares 

```python
data = pd.read_csv("data/FAOSTAT_data_en_10-28-2025_grouped.csv")
```

```python
data.head(5)
```

```python
data = data[data['Element'] == 'Gross Production Value (current thousand US$)']
```

```python
data.head(10)
```

```python
data.to_csv("data/production.csv", index=False)
```

## Tabla de enfermedades y padecimientos alimenticios

```python
df_obesidad = pd.read_csv("data/FAOSTAT_data_en_10-27-2025_obesidad_desnutricion.csv")
```

```python
df_obesidad.head(3)
```

```python
df_obesidad = df_obesidad[['Area', 'Item', 'Year', 'Unit', 'Value']]
df_obesidad.head(10)
```

```python
df_obesidad = df_obesidad[df_obesidad['Item'] == 'Number of obese adults (18 years and older) (million)']
df_obesidad.head(2)
```

```python
df_obesidad['Value'] = pd.to_numeric(df_obesidad['Value'], errors='coerce')
```

```python
df_obesidad['Value'] = df_obesidad['Value']*1000000
df_obesidad[['Area', 'Year', 'Value']]
df_obesidad.head()
```

```python
df_obesidad = df_obesidad[['Area', 'Year', 'Value']]
df_obesidad = df_obesidad[df_obesidad['Area'].isin(count_list)]
df_obesidad.head(20)
```

```python
len(df_obesidad)
```

```python
df_obesidad = df_obesidad[df_obesidad['Value'].notnull()]
```

```python
len(df_obesidad)
```
