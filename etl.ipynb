## Tabla de enfermedades y padecimientos alimenticios

```python
df = pd.read_csv("data/FAOSTAT_data_en_10-27-2025_obesidad_desnutricion.csv")
```

```python
df.head(3)
```

```python
df = df[['Area', 'Item', 'Year', 'Unit', 'Value']]
df.head(10)
```

```python
df_obesidad = df[df['Item'] == 'Number of obese adults (18 years and older) (million)']
df_obesidad['Value'] = pd.to_numeric(df_obesidad['Value'], errors='coerce')
df_obesidad['Value'] = df_obesidad['Value']*1000000
df_obesidad[['Area', 'Year', 'Unit', 'Value']]
df_obesidad.head()
```

```python
df_obesidad = df_obesidad[['Area', 'Year', 'Value']]
df_obesidad.head()
```

```python
len(df_obesidad)
```

```python
df_obesidad = df_obesidad[df_obesidad['Value'].notnull()]
```

```python
len(df_obesidad)
```

```python
df_obesidad.to_csv('data/df_obesidad.csv', index=False)
```
